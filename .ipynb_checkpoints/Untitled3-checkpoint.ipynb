```python
cat_var=['WorkClass','Educatoin','Marital_Status','Occupation','Sex','Native_country']
adult_var=adult.columns.values.tolist()
to_keep=[i for i in adult_var if i not in cat_var]
```

```python
# Our final data columns will be:

adult_final=adult[to_keep]
adult_final.columns.values
```

```python
adult_final=adult_final.drop(['Race'],axis=1)
```

```python
adult_final.head()
```

# Over-sampling using SMOTE

```python
adult_final['Income'].unique()
```

```python
# First replace the Income varible

adult_final['Income'].replace(' <=50K','0',inplace=True)
adult_final['Income'].replace(' >50K','1',inplace=True)
```

```python
X=adult_final.loc[:, adult_final.columns !='Income']
Y=adult_final.loc[:, adult_final.columns =='Income']
```

```python
X
```

```python
Y
```

```python
!pip install imbalanced-learn
```

```python
!pip install scikit-learn
```

```python
from imblearn.over_sampling import SMOTE

os = SMOTE(random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
columns = X_train.columns

os_data_X,os_data_y=os.fit_sample(X_train,y_train)
os_data_X=pd.DataFrame(os_data_X,columns=columns)
os_data_y=pd.DataFrame(os_data_y,columns=['Income'])
```

```python
# we can Check the numbers of our data
print("length of oversampled data is ",len(os_data_X))
print("Number of less income 50k in oversampled data",len(os_data_y[os_data_y['Income']==0]))
print("Number of more slary 50k",len(os_data_y[os_data_y['Income']==1]))
print("Proportion of less income 50k in oversampled data is ",len(os_data_y[os_data_y['Income']==0])/len(os_data_X))
print("Proportion of  more slary 50k data in oversampled data is ",len(os_data_y[os_data_y['Income']==1])/len(os_data_X))
```

# Recursive Feature Elimination

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
rfe = RFE(logreg, 90)
rfe=rfe.fit(os_data_X,os_data_y.values.ravel())
print(rfe.support_)
print(rfe.ranking_)
```

# Implementing the model

```python
col=['Capital_Gain','Capital_Loss','Hours_per_week']
x=os_data_X[col]
y=os_data_y['Income']
```

```python
x.dtypes
```

```python
y.dtypes
```

```python
import statsmodels.api as sm

logit_model=sm.Logit(y.astype(float), x.astype(float))
Result=logit_model.fit()
print(Result.summary())
```

# Logistic Regression Model Fitting

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

# Predicting the test set results and calculating the accuracy

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

# Confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
